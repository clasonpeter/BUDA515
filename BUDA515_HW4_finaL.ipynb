````python
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
````

````python
house_train = pd.read_csv("https://raw.githubusercontent.com/binbenliu/Teaching/main/data/BostonHousing_train.csv", header=0)
house_train
````

````python
house_train.columns
````

````python
house_test = pd.read_csv("https://raw.githubusercontent.com/binbenliu/Teaching/main/data/BostonHousing_test.csv", header=0)
house_test
````

````python
# training data
house_train_y = house_train['medv']
house_train_X = house_train.drop(labels = ['medv'], axis = 1)

# test data
house_test_y = house_test['medv']
house_test_X = house_test.drop(labels = ['medv'], axis = 1)
````

## (1) Specify and train your linear regression model

https://scikit-learn.org/1.5/modules/generated/sklearn.linear_model.LinearRegression.html

````python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(house_train_X, house_train_y)

print(f"Number of features: {house_train_X.shape[1]}")
print(f"Number of training samples: {house_train_X.shape[0]}")
````

## (2) predict and evaluate

````python
# your  code
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Make predictions on test data
house_test_y_pred = model.predict(house_test_X)

# Calculate evaluation metrics
r2 = r2_score(house_test_y, house_test_y_pred)
mae = mean_absolute_error(house_test_y, house_test_y_pred)
mse = mean_squared_error(house_test_y, house_test_y_pred)
rmse = np.sqrt(mse)

print("Model Evaluation on Test Data")
print(f"R² Score:                       {r2:.4f}")
print(f"Mean Absolute Error (MAE):      {mae:.4f}")
print(f"Mean Squared Error (MSE):       {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
````

## (3) Use the .coef_ to display the estimated coefficients for the linear regression model. Discuss how the different features affect the house prices.

````python
# Create a dataframe with feature names and coefficients
coefficients = pd.DataFrame({
    'Feature': house_train_X.columns,
    'Coefficient': model.coef_
})

print(coefficients)
print(f"\nIntercept: {model.intercept_:.4f}")
# Discussion
print("\nDiscussion:")
print("Positive coefficients mean the feature increases house price.")
print("Negative coefficients mean the feature decreases house price.")
print("Larger absolute values mean stronger impact on price.")
````

# (1) Read text data

The SMS Spam Collection v.1 is a set of SMS tagged messages that have been collected for SMS Spam research. It contains one set of SMS messages in English of 5,574 messages, tagged acording being ham (legitimate) or spam.

https://github.com/binbenliu/BUDA515/tree/main/smsspam

Training data: https://raw.githubusercontent.com/binbenliu/BUDA515/main/smsspam/SMSSpamCollection.train

Testing data: https://raw.githubusercontent.com/binbenliu/BUDA515/main/smsspam/SMSSpamCollection.test

````python
import pandas as pd

# read training data
train_file = "https://raw.githubusercontent.com/binbenliu/BUDA515/main/smsspam/SMSSpamCollection.train"
train_df = pd.read_csv(train_file, sep='\t', header=None, names=['label', 'text'])

print(f'num train records: {len(train_df)}')
train_df.head()
````

````python
# read test data
test_file = "https://raw.githubusercontent.com/binbenliu/BUDA515/main/smsspam/SMSSpamCollection.test"
test_df = pd.read_csv(test_file, sep='\t', header=None, names=['label', 'text'])

print(f'num test records: {len(test_df)}')
test_df.head()
````

###generate labels

````python
train_df['label'].replace(["ham","spam"], [0,1], inplace=True)
train_df.head()
````

````python
test_df['label'].replace(["ham","spam"], [0,1], inplace=True)
test_df.head()
````

````python
# train data
text_train = train_df['text']
y_train = train_df['label']

# test data
text_test = test_df['text']
y_test = test_df['label']
````

# (3) Specify and train your model (classifier)

You can use logistic regression, but there are many different models for classfication, so you can also try other methods.

**Regularziation**

\begin{equation}
L(\mathbf{w}, b) = \frac{1}{n}\sum_{i=1}^n l\left(\sigma\left(\mathbf{w}^\top \mathbf{x}^{(i)} + b\right),  y^{(i)}\right) + \frac{\alpha}{2} \|\mathbf{w}\|^2.
\end{equation}

**sklearn.linear_model.LogisticRegression**(penalty='l2', *, dual=False, tol=0.0001, C=1.0, fit_intercept=True, intercept_scaling=1, class_weight=None, random_state=None, solver='lbfgs', max_iter=100, multi_class='auto', verbose=0, warm_start=False, n_jobs=None, l1_ratio=None)

````python
# I will use CountVectorizer since it is simpler for me to understand
from sklearn.feature_extraction.text import CountVectorizer

# Create the vectorizer
vectorizer = CountVectorizer()

# Fit on the training messages and transform them into numbers
X_train = vectorizer.fit_transform(text_train)

# Transform the test messages (IMPORTANT: use the SAME vectorizer)
X_test = vectorizer.transform(text_test)

X_train.shape, X_test.shape
````

````python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)

model.fit(X_train, y_train)

print(f"Number of features: {X_train.shape[1]}")
print(f"Number of training samples: {X_train.shape[0]}")

````

#(4) predict and evaluate

Make predictions and get following classfication metrics:

* accuracy
* precision
* recall
* f1_score

You can also provide the confusion matrix as we used in the class.

````python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

````

````python
#  Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)   # spam = 1
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
````
